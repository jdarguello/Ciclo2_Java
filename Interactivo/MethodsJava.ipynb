## __1. Generalidades de un método__

Un método es la base de programación orientada a objetos (OOP, por sus siglas en inglés). Desde este punto, la programación adopta la forma de una "caja negra", en donde se recibe información de entrada que se procesa para generar una salida, como se aprecia en la Figura 2.

<div align="center">



<i>Figura 2.</i> Concepto de una función o _método_. Fuente: https://fhernanb.github.io/Manual-de-R/creafun.html

</div>




La estructura básica de un método es la siguiente:

```
<privacidad> <naturaleza> nombre_metodo (argumentos_entrada) {
  (desarrollo)
}
```

Un ejemplo con el que se convive diariamente en Java es el método _main_, que se trata del método principal de la clase de referencia. Presenta la siguiente estructura.

````
public static void main(String[] args) {
  System.out.println("Ejemplo del método principal de una clase");
}

main(new String[]{}); //Solicita como argumento de entrada un array tipo String.
                      //Se le envía un String vacío para que entre al método.
````

El método _main_ se compone de lo siguiente:

* `public` $\rightarrow$ significa que el método es de tipo _público_, lo que significa que puede ser llamado desde otra clase.
* `static` $\rightarrow$ característica que define un método o propiedad de una clase como _estático_. Es una instancia que globaliza el método, o propiedad, de una clase en todos los objetos creados a partir de esta (concepto mejor detallado en la sección de _Clases_).
* `void` $\rightarrow$ naturaleza del método. Este tipo de naturaleza no produce un retorno o salida.
* _main_ $\rightarrow$ nombre del método.
* `String[] args` $\rightarrow$ variable de entrada con nombre _args_. Se trata de un vector de tipo String.


## __2. Naturaleza de los métodos__

<div align="justify">

Además del método main, se pueden crear métodos que desarrollen una labor particular. Estos métodos se busca que retornen un valor con una naturaleza especificada. Esta naturaleza puede ser una _variable primitiva_, _clase contenedora_ o _arrays_, entro otros.

Todas las naturalezas de los métodos deben retornar un contenido del mismo tipo. Si el método es tipo `int`, el retorno también debe ser de tipo `int`. Sólo el tipo `void` está exento de retornar un valor.

</div>

Por ejemplo: creación de un método suma.

````
public int suma (int x ,int y) {
  return x+y;
}

System.out.println(suma(2,3));    //Suma de dos números. La naturaleza es de tipo int
````

````
//Es posible también emplear otro tipo de naturaleza. El ejemplo anterior fallará si se envían números decimales.
//Algo más "universal" sería:

public double suma (double x, double y) {
  return x+y;
}

System.out.println(suma(5.5, 2.3));
````

Es posible que la naturaleza de las entradas sea distinta a la de la función. 

### Ejemplo:

Elabora un algoritmo que identifique si una persona es mayor de edad.

````
public String mayoria_edad (String nombre, byte edad) {
  String mensaje = nombre + " tiene " + edad + " años,";
  String tipo;
  if (edad >= 18) {
    tipo = "mayor";
  } else {
    tipo = "menor";
  }
  mensaje += " por lo que es " + tipo + " de edad.";
  return mensaje;
}

Object[] persona = {"Alfredo", 22};
String respuesta = mayoria_edad((String)persona[0], (byte)((int)persona[1]));

System.out.println(respuesta);
````
